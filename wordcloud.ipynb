<b>WordCloud</b><br> WordCloud is used to visualize keywords, tags

Lets look at Penguin Movie reviews gathered from IMDB

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as npy
from PIL import Image
STOPWORDS.add('movie')
STOPWORDS.add('watch')
STOPWORDS.add('watching')
dataset = open("/home/labsuser/reviews_title.txt", "r").read()
def create_word_cloud(string):
    cloud = WordCloud(background_color = "white", max_words = 80, stopwords = set(STOPWORDS))
    cloud.generate(string)
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(cloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    plt.show()       
dataset = dataset.lower()
create_word_cloud(dataset)
```

```python
description = open("/home/labsuser/reviews_description.txt", "r").read()
create_word_cloud(description.lower())
```
